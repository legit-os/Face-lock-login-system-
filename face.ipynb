```python
from facenet_pytorch import MTCNN
device = "cpu"
import cv2
from PIL import Image
import numpy as np
```

```python
mtcnn = MTCNN(select_largest=True,)
```

```python
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    print(frame.shape)
    cropped, prob = mtcnn(frame.reshape((1,480,640,3)),return_prob=True)
    try:
        if cropped is not None:
            # print(len(cropped))
            # print(prob[0][0])
            # print(type(cropped[0]))
            cropped = cropped[0].numpy()
            # print(cropped.shape)
            cropped = np.transpose(cropped,(1,2,0))
            # print(cropped.shape)
            cropped = (cropped * 255).astype(np.uint8)
            # print(cropped)
        cv2.imshow('cam',cropped)
        # image = Image.fromarray(cropped)
    except Exception as e:
        # print(e)
        # break
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        continue
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    # break

cap.release()
cv2.destroyAllWindows()
```
